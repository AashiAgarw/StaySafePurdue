```python
from bokeh.models import ColumnDataSource

def plot(lat, lng, zoom=15, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    p = gmap(api_key, gmap_options, title='Crimes at Purdue', 
             width=bokeh_width, height=bokeh_height)
    # definition of the column data source: 
    print(type(df['lat'].dtype))
    source = ColumnDataSource(df.head(100))
    # see how we specify the x and y columns as strings, 
    # and how to declare as a source the ColumnDataSource:
    center = p.circle('lat', 'lon', size=10, alpha=0.5, 
                      color='yellow', source=source)
    show(p)
    
    return p

p = plot(lat, lon, map_type='terrain')
```

```python
def plot(lat, lng, zoom=15, map_type='hybrid'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    # the tools are defined below: 
    p = gmap(api_key, gmap_options, title='Crimes at Purdue', 
             width=bokeh_width, height=bokeh_height,
             tools=['hover', 'reset', 'wheel_zoom', 'pan'])
    source = ColumnDataSource(df)
    center = p.circle('lat', 'lon', size=9, alpha=0.5, 
                      color='yellow', source=source)
    show(p)
    return p

p = plot(lat, lon, map_type='hybrid', zoom=15)

```

```python
from bokeh.models import HoverTool

def plot(lat, lng, zoom=10, map_type='roadmap'):
    gmap_options = GMapOptions(lat=lat, lng=lng, 
                               map_type=map_type, zoom=zoom)
    # the tools are defined below: 
    hover = HoverTool(
        tooltips = [
            # @price refers to the price column
            # in the ColumnDataSource. 
            ('Nature', '@Nature'),
            #('General Location', '@General Location'),
        ]
    )
    # below we replaced 'hover' (the default hover tool), 
    # by our custom hover tool
    p = gmap(api_key, gmap_options, title='Crimes at Purdue', 
             width=bokeh_width, height=bokeh_height,
             tools=[hover, 'reset', 'wheel_zoom', 'pan'])
    source = ColumnDataSource(df)
    center = p.circle('lat', 'lon', size=15, alpha=0.2, 
                      color='magenta', source=source)
    show(p)
    return p

p = plot(lat, lon, map_type='roadmap', zoom=15)
```
